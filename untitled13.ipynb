```python
def year(y):
  df_year=df[df['Year']==y]
  year_tally=df_year.drop_duplicates(subset=['NOC','City','Team',	'Sport','Year','Event','Medal'])
  year_tally=year_tally.groupby('Team').sum()[['Gold','Silver','Bronze']].sort_values("Gold",ascending=False).reset_index()
  year_tally['total']=year_tally['Gold']+year_tally['Silver']+year_tally['Bronze']


  return year_tally.head(60)

```

```python
year(2016)
```

```python
def country(c):
  df_country=df[df['Team']==c]
  df_country=df_country.drop_duplicates(subset=['NOC','City','Team',	'Sport','Year','Event','Medal'])
  df_country=df_country.groupby('Year').sum()[['Gold','Silver','Bronze']].sort_values('Year').reset_index()
  df_country['total']=df_country['Gold']+df_country['Silver']+df_country['Bronze']
  return df_country
```

```python
country('India')
```

```python
def cy(c,y):
  df_cy=df[(df['Team']==c) & (df['Year']==y)]
  df_cy=df_cy.drop_duplicates(subset=['NOC','City','Team',	'Sport','Year','Event','Medal'])
  df_cy=df_cy.groupby('Team').sum()[['Gold','Silver','Bronze']].reset_index()
  df_cy['total']=df_cy['Gold']+df_cy['Silver']+df_cy['Bronze']
  return df_cy
```

```python
cy('India',2012)
```

```python
def sport_success(sport):
  df_sport=df[df['Sport']==sport]

  df_sport=df_sport.dropna()

  df_sport=df_sport.groupby('Name').sum()[['Gold','Silver','Bronze']].reset_index()
  df_sport['total']=df_sport['Gold']+df_sport['Silver']+df_sport['Bronze']
  df_sport=df_sport.sort_values('total',ascending=False)
  df_sport=df_sport
  df_sport = df_sport.merge(df[['Name', 'NOC']], on='Name', how='left')
  df_sport['NOC'] = df_sport['NOC'].fillna('none')



  return df_sport.head(1)
```

```python
sport_success('Badminton')
```
